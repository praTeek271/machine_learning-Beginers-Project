**1. Linear Regression**

```python
#code for linear regression.
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#Create the arrays that represent the values of the x and y axis:
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
#Execute a method that returns some important key values:
slope, intercept, r, p, std_err = stats.linregress(x, y)
```

```python
#A function is created to return new variable. 
def myfunc(x):
  return slope * x + intercept
```

```python
#Run each value of the x array through the function.
mymodel = list(map(myfunc, x))
```

```python
#Draw the original scatter plot,
#Line of polynomial regression 
#Displaying the diagram
plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

**2. Polynomial Regression**

```python
#code for polynomial regression
import numpy
import matplotlib.pyplot as plt
```

```python
#Create the arrays that represent the values of the x and y axis:
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
```

```python
#NumPy has a method that lets us make a polynomial model:
mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))
```

```python
#Then specify how the line will display, we start at position 1
#And end at position 22:
myline = numpy.linspace(1, 22, 100)
```

```python
#Draw the original scatter plot,
#Line of polynomial regression 
#Displaying the diagram
plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```
